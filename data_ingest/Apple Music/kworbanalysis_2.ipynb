This file is for analysis of the resulting file from kworbanalysis.ipynb.

```python
import pandas as pd

# Load the CSV file into a pandas dataframe
df = pd.read_csv('kworb_popularity_raw.csv', header=None, encoding='ISO-8859-1')
```

Column value at index 3 is currently in the following form:
"{'danceability': 0.575, 'energy': 0.33, 'key': 7, 'loudness': -6.557, 'mode': 1, 'speechiness': 0.031, 'acousticness': 0.416, 'instrumentalness': 0, 'liveness': 0.0872, 'valence': 0.277, 'tempo': 95.801, 'type': 'audio_features', 'id': '6QfS2wq5sSC1xAJCQsTSlj', 'uri': 'spotify:track:6QfS2wq5sSC1xAJCQsTSlj', 'track_href': 'https://api.spotify.com/v1/tracks/6QfS2wq5sSC1xAJCQsTSlj', 'analysis_url': 'https://api.spotify.com/v1/audio-analysis/6QfS2wq5sSC1xAJCQsTSlj', 'duration_ms': 217213, 'time_signature': 4}"

We want to split this up into multiple columns.

```python
# Convert the string dictionary to a Python dictionary
df[3] = df[3].apply(lambda x: eval(x))

# Use json_normalize() to expand the dictionary into multiple columns
df_normalized = pd.json_normalize(df[3])

# Concatenate the original DataFrame with the new DataFrame containing the normalized dictionary columns
df = pd.concat([df, df_normalized], axis=1)
```

```python
# Drop the original column with the string dictionary
df.drop(df.columns[3], axis=1, inplace=True)

# Rename unnamed columns
df.rename(columns={0: 'popularity_rank', 1: 'title', 2: 'popularity'}, inplace=True)

# Drop columns that are useful
df.drop(['track_href', 'analysis_url', 'type'], axis=1, inplace=True)

# Cast numeric types to numeric
numeric_cols = ['popularity_rank', 'popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
                'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Save the dataframe to a CSV file
df.to_csv('kworb_popularity.csv', index=False)
```
